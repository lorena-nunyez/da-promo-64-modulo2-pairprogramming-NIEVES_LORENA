````python
import requests
````

````python
API_KEY = {'Authorization': '773999c8-a2d9-4e4e-92d4-83fc1f7e338f'}

url = 'https://api.balldontlie.io/v1/players'

params1 = {
    "cursor": 0,
    "per_page": 100
  }
````

````python
respuesta1 = requests.get(url, headers=API_KEY, params=params1)

if respuesta1.status_code == 200:
    print(respuesta1.content)
    print(respuesta1.headers)
else:
    print(f'Error: EL código de error es {respuesta1.status_code}')
````

````python
params2 = {
    "cursor": 100,
    "per_page": 100
  }

respuesta2 = requests.get(url, headers=API_KEY, params=params2)

if respuesta2.status_code == 200:
    print(respuesta2.content)
    print(respuesta2.headers)
else:
    print(f'Error: EL código de error es {respuesta2.status_code}')
````

````python
jugadores1 = respuesta1.json()
jugadores1
````

````python
jugadores2 = respuesta2.json()
jugadores2
````

````python
jugadores_total = jugadores1['data'] + jugadores2['data']
jugadores_total
````

````python
len(jugadores_total)
````

````python
jugadores_total[0].keys()
````

````python
tabla_jugadores = {'Nombre': [], 'Apellidos': [], 'Equipo': []}

for jugador in jugadores_total:
    tabla_jugadores['Nombre'].append(jugador['first_name'])
    tabla_jugadores['Apellidos'].append(jugador['last_name'])
    tabla_jugadores['Equipo'].append(jugador['team']['full_name'])

print(tabla_jugadores)
    
````

````python
import pandas as pd

df_tabla_jugadores = pd.DataFrame(tabla_jugadores)

df_tabla_jugadores
````
