**Buscar un personaje de Star Wars**

Ahora que conocemos mejor el API de Star Wars, vamos a crear una aplicación en Python que la utilice. El programa pedirá al usuario, mediante un input(), que escriba el nombre (o parte del nombre) de un personaje. A continuación, hará una búsqueda en el API de Star Wars y mostrará por consola un listado con los personajes que coinciden con la búsqueda, indicando su nombre y su género.

```python
def mostrar_personaje(nombre):
    url = 'https://swapi.py4e.com/api/people/'
    respuesta = requests.get(url)
    if respuesta.status_code == 200:
        personajes = respuesta.json()
        for personaje in personajes['results']:
            if personaje['name'] == nombre:
                print(f'El personaje {personaje['name']} es del género {personaje['gender']}')
                return
        print('Personaje no encontado')
    else:
        print(f'Error: {respuesta.status_code}')
        
nombre = input('Introduce el personaje a buscar')
mostrar_personaje(nombre)
```
